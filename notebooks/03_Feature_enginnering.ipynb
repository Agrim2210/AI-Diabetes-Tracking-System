```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:\\Users\\HP\\Desktop\\Diabetes-Prediction-System\\data\\diabetes_clean.csv")
df.head()
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["BMI_Category"] = df["BMI"].apply(bmi_category)

df[["BMI", "BMI_Category"]].head()
```

```python
def age_group(age):
    if age < 30:
        return "Young"
    elif age < 50:
        return "Middle"
    else:
        return "Old"

df["Age_Group"] = df["Age"].apply(age_group)

df[["Age", "Age_Group"]].head()
```

```python
def glucose_category(glucose):
    if glucose < 100:
        return "Normal"
    elif glucose < 140:
        return "Prediabetes"
    else:
        return "Diabetes"

df["Glucose_Level"] = df["Glucose"].apply(glucose_category)

df[["Glucose", "Glucose_Level"]].head()
```

```python
df["BMI_Age_Interaction"] = df["BMI"] * df["Age"]

df[["BMI", "Age", "BMI_Age_Interaction"]].head()
```

##  Feature Engineering Summary

- Created BMI categories based on health standards
- Grouped age into meaningful life stages
- Categorized glucose levels based on medical thresholds
- Added interaction feature (BMI × Age)

These transformations improve interpretability and model learning.
