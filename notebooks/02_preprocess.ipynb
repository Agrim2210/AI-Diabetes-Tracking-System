```python
import pandas as pd
import numpy as np 
df=pd.read_csv("c:\\Users\\HP\\Desktop\\Diabetes-Prediction-System\\data\\diabetes_raw.csv")
df.head()
```

```python
zero_cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[zero_cols]=df[zero_cols].replace(0,np.nan)
df.fillna(df.median(),inplace=True)
```

```python
for col in df.columns[:-1] :
    
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    lr=q1-iqr*1.5
    ub=q3+1.5*iqr
    df[col]=df[col].clip(lower=lr,upper=ub)    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for cols in df.columns:
    sns.boxplot(df[cols])
    plt.title(cols)
    plt.show()
# Hence outliers from every column is removed     
```

```python
df.to_csv("C:\\Users\HP\\Desktop\\Diabetes-Prediction-System\\data\\diabetes_clean.csv",index=False)
```
